## Part 3: Analyzing the Data

### 3.1: Monthly Temperatures

Here, I want to analyze the values of my monthly temperatures. First, I'll separate my temperature values by month and find the mean temperature for each month.

```python
monthly_avg = pd.DataFrame(df['temp_f'].groupby(df.date_time.dt.month).mean())
```

```python
monthly_avg
```

Each of my months has been grouped into my index - I want to make a column for each month so it's easier to manipulate.

```python
monthly_avg['month'] = monthly_avg.index
```

```python
monthly_avg
```

To make this easier to read, I'm going to switch the position of my month and temperature columns.

```python
monthly_avg = monthly_avg[["month","temp_f"]]
```

```python
monthly_avg
```

I also want to get rid of the index name so my data is easier to read.

```python
monthly_avg.index.name = None
```

```python
monthly_avg
```

Now, I want to change my month representation to the string of the actual month name. To do this, I will be using a dictionary and passing it into a function that makes a list that I can pass as my new column values.

```python
month_dict = {1:"January",
             2:"February",
             3:"March",
             4:"April",
             5:"May",
             6:"June",
             7:"July",
             8:"August",
             9:"September",
             10:"October",
             11:"November",
             12:"December"}
```

```python
def convertList(dictName, convertColumn):
    listIndex = list(convertColumn)
    convertedIndex = list(map(lambda x: dictName.get(x), listIndex))
    return convertedIndex
```

```python
monthly_avg['month'] = convertList(month_dict, monthly_avg['month'])
```

```python
monthly_avg
```

### 3.2: Hourly Temperatures

Now, suppose I want to analyze the average temperature for every hour across all months.

```python
hourly_avg = df['temp_f'].groupby(df.date_time.dt.hour).mean()
```

```python
hourly_avg
```

Since my index value represents the hour in military time, I'm going to change the index name from date_time (which is the column I pulled the time from) to Hour.

```python
hourly_avg.index.name = "Hour"
```

```python
hourly_avg
```

Now, I also want to convert my times from military time to standard time. I will do this using the same function I used to convert my months.

```python
hour_dict = {0:"12:00AM",
            1:"1:00AM",
            2:"2:00AM",
            3:"3:00AM",
            4:"4:00AM",
            5:"5:00AM",
            6:"6:00AM",
            7:"7:00AM",
            8:"8:00AM",
            9:"9:00AM",
            10:"10:00AM",
            11:"11:00AM",
            12:"12:00PM",
            13:"1:00PM",
            14:"2:00PM",
            15:"3:00PM",
            16:"4:00PM",
            17:"5:00PM",
            18:"6:00PM",
            19:"7:00PM",
            20:"8:00PM",
            21:"9:00PM",
            22:"10:00PM",
            23:"11:00PM"}
```

```python
hourly_avg.index = convertList(hour_dict, hourly_avg.index)
```

```python
hourly_avg
```

### 3.3: Weather Types

Finally, I want to guage the various types of weather that have been recorded.

```python
df['weather'].groupby(df['weather']).count()
```

Above we were able to condense the weather data into the different values that were used. However, notice that some rows combine multiple values which actually ends up reusing some values multiple times.

To help with this, I'm going to convert my weather values to a dataframe, move them to a column, then separate the values so that I can condense my list so it doesn't have repeating values.

```python
weather = pd.DataFrame(df['weather'].groupby(df['weather']).count())
```

```python
weather
```

```python
weather['weatherType'] = weather.index
```

```python
weather
```

```python
weather[['Weather 1', 'Weather 2', 'Weather 3']] = weather.weatherType.str.split(",", expand = True)
```

```python
weather
```

```python
uniqueWeather = list(pd.unique(weather[['Weather 1', 'Weather 2', 'Weather 3']].values.ravel()))
```

```python
uniqueWeather
```

An unfortunate circumstance of separating my data the way I did was that I had some N/A values since each data value didn't have the same amount of delimited values. To manipulate this data further, I want to get rid of those nonexistent values.
